```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Obengo el html
url = 'https://practicum-content.s3.us-west-1.amazonaws.com/data-analyst-eng/moved_chicago_weather_2017.html'
req = requests.get(url).text

# Creo el objeto soup para poder manipular los datos
soup = BeautifulSoup(req, 'lxml')
```

```python
# Identifico y extraigo la tabla de la pag web
table = soup.find('table', attrs = {'id': 'weather_records'})
```

```python
# Comienzo a crear el DataFrame
# Lista donde se almacenaran todos los nombres de columnas
heading_table = []

# Iterar sobre cada fila de la tabla para extraer unicamente los nombres de las columnas
for row in table.find_all('th'):
    heading_table.append(row.text)
```

```python
# Lista donde se almacenarán los datos de la tabla
content = []

# Iterar cada fila para extraer todos los registros (filas) de la tabla
# Cada fila está envuelta en una etiqueta <tr>
for row in table.find_all('tr'):
# Esta condicion ignora la primera fila, los encabezados
    if not row.find_all('th'):
# Dentro de cada fila, el contenido de la celda está envuelto en etiquetas <td> </td>
# Necesitamos recorrer todos los elementos <td>, extraer el contenido de las celdas y agregarlo a la lista
# Luego agregamos cada una de las listas a la lista de contenido
        content.append([element.text for element in row.find_all('td')])
```

```python
# paso la lista de contenido como datos y heading_table como encabezados
weather_records = pd.DataFrame(content, columns = heading_table)

print(weather_records.head())
```
